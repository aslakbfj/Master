```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
#add the path to the metrics_csv file
sys.path.append('C:/Users/aslak/Master/Github_first/Master/data/salmon')
```

```python

# Load the metrics into a pandas DataFrame
df = pd.read_csv("metrics.csv")
df = df.dropna(how='all')

# sort the data with respect to the num_units
df = df.sort_values(by=['num_units'])
    # Create a new figure
plt.figure()

    # Plot AUCROC
plt.plot(df['num_units'], df['AUCROC'], label='AUCROC')

# Plot AUCPR
plt.plot(df['num_units'], df['AUCPR'], label='AUCPR')
#plt.xlim(0,500)
#plt.ylim(0,1)
    # Add a legend
plt.legend()

    # Add labels
plt.xlabel('num_units')
plt.ylabel('Score')

plt.tight_layout()
plt.show()
```
